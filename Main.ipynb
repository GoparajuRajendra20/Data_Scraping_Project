```python
class MainProcess:
    def __init__(self):
        # Define URLs and file paths
        self.api_url = 'https://medium.com/'
        self.website_url = 'https://medium.com/'
        self.pdf_path = 'C:\\Rajendra_OS_D\\Downloads\\Resume - Rajendra Goparaju Data Scrapping Assistant.pdf'

    def execute(self):
        # Step 1: API Requests
        api_data = {}
        api_data = fetch_data_from_api(self.api_url)
        print("Fetched the data by interacting the REST API using requests library.")
        for key, value in api_data.items():
            print(f"{key}: {value}")
        print("\n")
        print("-------------------------------------------------------")
        print("\n")
            

       
        # Step 2: Data Scraping
        scraped_data = scrape_data_from_website(self.website_url)

        print("Performed data scrapping by retrieving the data of interest from the medium website including titles, sub_titles, source_urls.")
        print("\n")
        for key, value in scraped_data.items():
            print(f"{key}: {value}")
        print("-------------------------------------------------------")
        print("\n")


        # Step 3: PDF Extraction
        pdf_text = extract_text_from_pdf(self.pdf_path)
        print("Conducted some basic text search and retrieved the data of interest leveraging PyMuPDF library.")
        print("\n")
        for key, value in pdf_text.items():
            print(f"{key}: {value}")

if __name__ == "__main__":
    main_process = MainProcess()
    result = main_process.execute()
```
